### Jointplot
It finds the relationship between two numeric variable
It is usually a scatter plot

```python
sns.jointplot(x='tip',y='total_bill', data=df,kind='hex')
```

### Pair plot
Also known as scatterplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker') 
```

### Dist Plot
It is used to create Histogram

```python
sns.distplot(df['tip'],kde=False,bins=10)
```
